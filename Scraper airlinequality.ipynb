```python
# import requests, time and beautifulsoup
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
# request access to the website and set up soup variable
url = 'https://www.airlinequality.com/airline-reviews/klm-royal-dutch-airlines/?sortby=post_date%3ADesc&pagesize=100'
res = requests.get(url)
review_source_code = res.text
soup = BeautifulSoup(review_source_code)
```

```python
# generating all 45 page urls of the KLM review page in a list
base_url = "https://www.airlinequality.com/airline-reviews/klm-royal-dutch-airlines/page/"
quote_page_urls = []

for counter in range(1, 45):
    full_url = base_url + str(counter)
    quote_page_urls.append(full_url)
    sleep(1)

print(quote_page_urls)
```

```python
# the firt hundred review titles
# 1 entry is wrong: "KLM Royal Dutch Airlines customer review"
counter = 1
while counter < 101:
    print(soup.find_all('h2')[counter].get_text())
    counter += 1
    sleep(1)
```

```python
# The first hundred review ratings
counter = 1
while counter < 101:
    print(soup.find_all(class_="rating-10")[counter].find("span").get_text())
    counter += 1
    sleep(1)
```

```python
# The first hundred review ratings in a list
review_ratings = []

counter = 1
while counter < 101:
    review_rating = soup.find_all(class_="rating-10")[counter].find("span").get_text()
    counter += 1
    review_ratings.append(review_rating)
    
print(review_ratings)
```

```python
# The first hundred review dates
counter = 0
while counter < 100:
    print(soup.find_all(class_="text_sub_header userStatusWrapper")[counter].find("time").get_text())
    counter += 1
```

```python
# The first hundred review dates in a different format
for link in soup.find_all("time"): 
    print(link.attrs["datetime"])
```

```python
# If the first hundred reviews are verified or not
counter = 0
while counter < 100:
    print(soup.find_all(class_="body")[counter].find("em").get_text())
    counter += 1
```
